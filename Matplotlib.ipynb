# **Matplotlib**

**Data Visualization** :--> Data visualization is the process of representing data in visual or graphical format.

such as charts, graphs, maps, and dashboard - to make information easier to understand and analyze.

it helps you see patterns, trends and relationship in data that might not be obvious in raw number or text .

**Common types of Data Visualization**



1. Bar Chart--> Compares values across (Categories)
2. Line Plot--> Shows trends over time
3. Pie Chart--> Shows parts of a whole (Categories)
4. Histogram--> Shows frequency destribution
5. Scatter Plot--> Show relationship between Two variables
6. Heatmap--> Shows data intensity using colors.

**Top 5 Advantage of  Data Visualization.**
1. Saves Time ---------------> Understand data quickly.
2. Easy Communication -------> Explain data to anyone.
3. Spat Trends and Patterns--> See data's movements.
4. Better Decision making----> Make informed choices.
5. Handles Big Data well-----> Understand thousands of rows.

```python


# Sample data
months = [1, 2, 3, 4]
sales = [1200, 1500, 1100, 1800]

# IMPORTANT: You must add the 'label' argument here
plt.plot(months, sales, label='Revenue')

plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title('Monthly Sales Data Report')

# Now this line will work because it finds 'Revenue'
plt.legend(loc='upper left', fontsize=12)

plt.grid(color='gray', linestyle=':', linewidth=1)
plt.xlim(1, 4)
plt.ylim(0, 2000)
plt.xticks([1, 2, 3, 4], ['M1', 'M2', 'M3', 'M4'])

plt.show()
```

**EX of Bar Chart**



```python
# plt.bar(x,y,hight,color='color_name',width=value,label='labelname')
product=['A','B','C','D']
Sales=[1000,1500,800,1200]

plt.bar(product,sales,color='blue',label='Sales 2025',linewidth=2)
plt.xlabel("product")
plt.ylabel("sales")
plt.title("product sales comparsion")
plt.legend()
plt.show()
```

**EX of Pie Chart**

```python
# plt.pie(values,label=lebel_list,outopct='%1.1f%%')

regions=['North','South','East','West']
revenue=[3000,2000,500,1000]
```

```python


plt.pie(revenue,
        labels=regions,
        autopct='%1.1f%%',
        colors=['gold', 'skyblue', 'lightgreen', 'coral'])

plt.title('Revenue contribution by Region')
plt.show()
```

 **EX of Histrogram**

```python
# plt.hist(data,bins=number_of_bins,colors='color_name',edgecolor='black')
scores=[45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,68,77,60]
```

**EX of Scatter graph**

```python
# plt.scatter(x,y,color='color_name',marker='maker_style',label='label_name)

hours_studies=[1,2,3,4,5,6,7,8]
exam_score=[50,55,60,65,70,75,80,85]
```

```python
plt.scatter(hours_studies,exam_score,color='green',marker='o',label='student Data')
plt.xlabel('Hours of studied')
plt.ylabel('exam score')
plt.title('Relationship between study time and exam score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter([1,2,3],[50,55,60],color='blue',label='class A') #g1
plt.scatter([1,2,3],[45,50,52],color='orange',label='class B') #g2

# Scatter plot
plt.xlabel('hours studie')
plt.ylabel('exam score')
plt.title('compersion of two classed')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# plt.subplot(nrows, ncols, index)
x=[1,2,3,4]
y=[10,20,15,25]
```

```python
plt.subplot(1,2,1) # 1 row, 2 column, 2nd suplot
plt.plot(x,y)
plt.title('Line Chart')


plt.subplot(1,2,2) # 1 row, 2 column, 2nd suplot
plt.bar(x,y)
plt.title('Bar Chart')


plt.tight_layout()
plt.show()
```

```python
# fig, ax=plt.subpots(nrows,ncols,figsize=(width, hight))
fig,ax=plt.subplots(1,2,figsize=(10,5))
x=[1,2,3,4]
y=[10,20,15,25]

ax[0].plot(x,y,color='blue')
ax[0].set_title('Line Chart')

ax[1].bar(x,y,color='green')
ax[1].set_title('Bar Chart ')

fig.suptitle('Comparsion of Line and Bar charts')

plt.tight_layout()
plt.show()
```
