Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure :[restaurent_id,2021,2022,2023,2024]
sales_data=np.array([
    [1, 150000, 180000, 220000, 250000],# Paradise Biryani
    [2, 120000, 140000, 160000, 190000],# Beijing Bites
    [3, 200000, 230000, 260000, 300000],#Pizza Hub
    [4, 180000, 210000, 240000, 270000],#Burger point
    [5, 160000, 185000, 205000, 230000],#chai point
    
])
```

```python
print("==== Zomoto sales analysys ====")
print("\n Sales data shap",sales_data.shape)
# sample data for 1st 3 restau/;
print("Sample data for 1st 3 restau:\n",sales_data[:3])
print("*************************\n",sales_data[:,1:])
```

########## Data Visualization ############

```python
# total sales per year\
print("# total sales per year")
print(np.sum(sales_data,axis=0))
yeary_total = np.sum(sales_data[:, 1:], axis=0)
print(yeary_total)
```

```python
# minimum sales per restaurent
print("# minimum sales per restaurent")
min_sales=np.min(sales_data[:,1:],axis=1)
print(min_sales)
```

```python
# maximum sales per year
print("# maximum sales per year")
max_sales=np.max(sales_data[:,1:],axis=0)
print(max_sales)
```

```python
# Average sales per restaurent
print("# Average sales per restaurent")
avg_sales =np.mean(sales_data[:, 1:],axis=1)
print(avg_sales)
```

```python
# Cumulative Cells(step by step sum of per year sales data)
print("# Cumulative Cells(step by step sum of per year sales data)")
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)
```

```python
# monthly average

monthly_avg = sales_data[:,1:] / 12 # brodcasting
print(monthly_avg)
```

Using Matplot


```python
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales accross all restaurent")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1=np.array([1,2,3,4,5])
vector2=np.array([6,7,8,9,10])
print("Vector addition:",vector1 + vector2)

print("\n Multiplication vector",vector1 * vector2)

print("\nDot produt",np.dot(vector1,vector2))

#  sim(a,b)= a.b/(||a||.||b||)

angle =np.arccos(np.dot(vector1,vector2)/ (np.linalg.norm(vector1)
*np.linalg.norm(vector2)))
print("\nangle",angle)
```
