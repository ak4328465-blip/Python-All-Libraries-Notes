### Array Reshaping

```python
arr = np.arange(12)
print("original array",arr)

reshaped=arr.reshape((3,4))
print("\nReshaped array\n",reshaped)
# print("size",reshaped.size)
# print("Dimension",reshaped.ndim)
# print("shap",reshaped.shape)
flatend=reshaped.flatten()
print("\n Flattened array",flatend)

# ravel (return view,instend of copy)
raveled=reshaped.ravel()
print("\n raveled array",raveled)
```

### Transpose of matrix



```python
transpose=reshaped.T
print("\n Transpose array\n",transpose)
print("size",transpose.size)
print("Dimension",transpose.ndim)
print("shap",transpose.shape)
```
